```python
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
import selenium.webdriver.support.expected_conditions as EC
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
```

```python
driver=webdriver.Chrome("C:\Program Files (x86)\Google\Chrome\Application\chromedriver.exe")
```

```python
driver.get('https://www.bilibili.com')
```

```python
input = WebDriverWait(driver, 10).until(EC.presence_of_element_located(
    (By.CSS_SELECTOR, '#nav-searchform > div.nav-search-content > input')))
```

```python
search_btn = WebDriverWait(driver, 10).until(EC.presence_of_element_located(
    (By.CSS_SELECTOR, '#nav-searchform > div.nav-search-btn')))
```

```python
input.send_keys('蔡徐坤 篮球')
search_btn.click()
```

```python
all_tag=driver.window_handles
driver.switch_to.window(all_tag[1])
```

```python
total_pages = WebDriverWait(driver, 10).until(EC.presence_of_element_located(
    (By.CSS_SELECTOR, '#i_cecream > div:nth-child(1) > div:nth-child(1) > div.search-content > div > div > div.flex_center.mt_x50.mb_x50 > div > div > button:nth-child(10)')))
total_pages = int(total_pages.text)
total_pages
```

```python
lines = [['标题', '播放量']]
for i in range(total_pages-1):
    html = driver.page_source
    # with open('tmp'+str(i)+'.html', 'w',encoding='utf-8') as f:
    #     f.write(html)
    soup = BeautifulSoup(html, 'lxml')
    list = soup.find_all(class_='bili-video-card')
    # print(list)
    for l in list:
        # print(l)
        title = l.select('div.bili-video-card__wrap.__scale-wrap > div > div > a > h3')
        if title==[]:
            continue
        title=title[0].get_text()
        print(title)
        played = l.select(
            'div.bili-video-card__wrap.__scale-wrap > a > div > div.bili-video-card__mask > div > div > span:nth-child(1) > span')
        played = played[0].get_text()
        print(played)
        lines.append([title, played])
    next_page_btn = WebDriverWait(driver, 10).until(EC.presence_of_element_located(
        (By.CSS_SELECTOR, 'div > div > button:nth-child(11)')))
    next_page_btn.click()
```

```python
print(lines[:3])
```

```python
import xlwt

book=xlwt.Workbook(encoding='utf-8',style_compression=0)
sheet=book.add_sheet('蔡徐坤',cell_overwrite_ok=True)

for i,line in enumerate(lines):
    sheet.write(i,0,line[0])
    sheet.write(i,1,line[1])

book.save('蔡徐坤和篮球.xlsx')
```
