```python
import requests
import time
from bs4 import BeautifulSoup
```

```python
import string

to_save=[]
for i in range(1,26):
    url='http://bang.dangdang.com/books/fivestars/01.00.00.00.00.00-recent30-0-0-1-'+str(i)
    print(url)
    try:
        time.sleep(2)
        response=requests.get(url)
        print(response.status_code)
        if response.status_code == 200:
            html=response.text
            # print(html)
            soup=BeautifulSoup(html,'lxml')
            res_list=soup.find(class_="bang_list clearfix bang_list_mode").find_all('li')
            # print(res_list)
            for r in res_list:
                r_range=r.find(class_="list_num red") if r.find(class_="list_num red")!=None else r.find(class_="list_num")
                r_range=r_range.string
                r_name=r.find(class_="name").find('a').text
                r_tuijian=r.find(class_="tuijian").string
                if r.find(class_="publisher_info").find('a') !=None:
                    r_author=r.find(class_="publisher_info").find('a').string
                else:
                    r_author='unknown'
                r_price=r.find(class_="price_r").string
                print("排名：",r_range,'|',"书名：",r_name,'|',"推荐：",r_tuijian,'|',"作者：",r_author,'|',"价格：",r_price)
                to_save.append([r_range,r_name,r_tuijian,r_author,r_price])
            # break
        
    except requests.RequestException:
        print('requests error')

```

```python
# print(to_save)

import xlwt

movies=xlwt.Workbook(encoding='utf-8',style_compression=0)

sheet=movies.add_sheet('当当Top500',cell_overwrite_ok=True)
sheet.write(0,0,'排名')
sheet.write(0,1,'书名')
sheet.write(0,2,'推荐')
sheet.write(0,3,'作者')
sheet.write(0,4,'价格')

for i,t in enumerate(to_save):
    for j in range(5):
        sheet.write(i+1,j,t[j])

movies.save(u'当当Top500图书.xlsx')
```
